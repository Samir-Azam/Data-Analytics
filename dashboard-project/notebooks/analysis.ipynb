```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("../data/raw/data.csv")
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")

df = df.rename(columns={
    "On_a_scale_of_1_to_5_,_how_often_do_you_shop_in_the_selected_product_category_monthly?": "Shopping_Frequency"
})
```

```python
df = df.rename(columns={
    "Say_for_example_,_a_similar_product_with_similar_features_is_being_sold_at_half_the_price_by_an_unknown_seller_as_compared_to_a_branded_product._Would_you_purchase_that_product?_or_Simply_Do_you_prefer_cost_over_brands?": "Prefer_Cheap_Product"
})
```

```python
df["Prefer_Cheap_Product"].value_counts()
```

```python
print("📊 Average Spending:")
print(df["Avg_Spend"].value_counts())

print("\n📊 Shopping Frequency:")
print(df["Shopping_Frequency"].value_counts())

print("\n📊 Product Category:")
print(df["Product_Category"].value_counts())

print("\n📊 Brand Preference:")
print(df["Prefer_Cheap_Product"].value_counts())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="Product_Category", data=df)
plt.title("Most Popular Product Categories")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Avg_Spend", data=df)
plt.xticks(rotation=45)
plt.title("Spending Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Shopping_Frequency", data=df)
plt.title("Shopping Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Prefer_Cheap_Product", data=df)
plt.title("Brand vs Cheap Preference")
plt.show()
```

```python
print("🔍 Key Insights:")
print("1. Clothing is the most popular category")
print("2. Most users spend between ₹1000–₹5000")
print("3. Users shop at moderate frequency")
print("4. Majority prefer branded products over cheaper alternatives")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Avg_Spend", y="Age", data=df)
plt.xticks(rotation=45)
plt.title("Age vs Spending")
plt.show()
```
