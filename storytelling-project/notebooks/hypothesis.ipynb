```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load data
df = pd.read_csv("../data/data.csv")

# Clean columns
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Rename long column
df = df.rename(columns={
    "Say_for_example_,_a_similar_product_with_similar_features_is_being_sold_at_half_the_price_by_an_unknown_seller_as_compared_to_a_branded_product._Would_you_purchase_that_product?_or_Simply_Do_you_prefer_cost_over_brands?": "Prefer_Cheap_Product"
})

# -------------------------------
# Create contingency table
# -------------------------------
table = pd.crosstab(df["Prefer_Cheap_Product"], df["Avg_Spend"])

print("Contingency Table:\n")
print(table)

# -------------------------------
# Chi-Square Test
# -------------------------------
chi2, p_value, dof, expected = chi2_contingency(table)

print("\nChi-Square Value:", chi2)
print("P-Value:", p_value)

# -------------------------------
# Interpretation
# -------------------------------
alpha = 0.05

if p_value < alpha:
    print("\n✅ Statistically Significant Relationship")
    print("👉 Spending pattern depends on brand preference")
else:
    print("\n❌ No Significant Relationship")
    print("👉 Spending pattern is independent of brand preference")
```
